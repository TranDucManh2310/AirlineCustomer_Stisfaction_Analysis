Thực hiện trên dữ liệu gốc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
df_preprocessed_ = pd.read_csv('../data/preprocessed_data (1).csv')
df_preprocessed_
```

```python
# Bỏ cột "Index" và "id"
df_preprocessed_ = df_preprocessed_.drop(columns=['Index', 'id'])
```

```python
X_ = df_preprocessed_.drop(columns=['satisfaction_satisfied'])
y_ = df_preprocessed_['satisfaction_satisfied'].values.ravel()
X_ = X_.to_numpy()
```

```python
# Chia tập train - test lần lượt theo các tỉ lệ: 6:4, 7:3, 8:2
from sklearn.model_selection import train_test_split

X_train_origin_82_, X_test_origin_82_, y_train_origin_82_, y_test_origin_82_ = train_test_split(X_, y_, test_size=0.2) #8:2
X_train_origin_73_, X_test_origin_73_, y_train_origin_73_, y_test_origin_73_ = train_test_split(X_, y_, test_size=0.3) #7:3
X_train_origin_64_, X_test_origin_64_, y_train_origin_64_, y_test_origin_64_ = train_test_split(X_, y_, test_size=0.4) #6:4
```

```python
X_train_origin_82_.shape
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

Tập dữ liệu gốc chia train-test: 8:2

```python
# Gọi mô hình và khớp dữ liệu trainning
softmax_model_82 = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=400)
softmax_model_82.fit(X_train_origin_82_, y_train_origin_82_)
```

```python
# Dự đoán xác suất cho hai phân lớp
y_train_origin_82 = softmax_model_82.predict_proba(X_train_origin_82_)
y_test_origin_82 = softmax_model_82.predict_proba(X_test_origin_82_)
```

```python
# Chia nhãn theo từng phân lớp
y_train_origin_82_0 = y_train_origin_82[:, 0]
y_train_origin_82_1 = y_train_origin_82[:, 1]
y_test_origin_82_0 = y_test_origin_82[:, 0]
y_test_origin_82_1 = y_test_origin_82[:, 1]
```

```python
# Dự đoán trên dữ liệu test lớp 0
y_pred_820 = linear_model_820.predict(X_test_origin_82_)
```

```python
# Tính toán các chỉ số hồi quy
print("\n--- Evaluation as Regression on class 0---")
print("MSE:", mean_squared_error(y_test_origin_82_0, y_pred_820))
print("MAE:", mean_absolute_error(y_test_origin_82_0, y_pred_820))
print("R²:", r2_score(y_test_origin_82_0, y_pred_820))
```
